```
import rsa
```

Alice cria a mensagem 𝑚 e calcula o hash H(𝑚)

```
message = 'Ola meu caro Bob'.encode()
hash = rsa.compute_hash(message, 'SHA-1')
```

```
print(hash)
```

Alice anexa H(𝑚) à mensagem 𝑚, criando uma mensagem estendida (𝑚, H(𝑚)), e a envia para Bob

Bob recebe uma mensagem estendida (𝑚, ℎ) e calcula H(𝑚). Se H(𝑚) = ℎ, Bob conclui que está tudo certo.



```
h = rsa.compute_hash(message, 'SHA-1')
if (h == hash):
  print(True)
else: 
  print(False)
```

Problema: um invasor poderia se passar por Alice e criar uma mensagem 𝑚’, calcular H(𝑚’) e enviar a bob.

Alice cria a mensagem 𝑚, concatena 𝑠 com 𝑚 e cria 𝑠+𝑚, e calcula o hash H(𝑠+𝑚). Alice anexa o MAC à mensagem 𝑚 (𝑚, H(𝑚+𝑠)).


```
s = rsa.compute_hash("segredo".encode(), 'SHA-1')
message = 'Ola meu caro Bob'.encode()
m = s + message
print(m)
```

```
hash = rsa.compute_hash(m, 'SHA-1')
print(hash)
```

Bob recebe uma mensagem estendida (𝑚, ℎ) e conhecendo 𝑠, calcula o MAC H(𝑚+𝑠). Se H(𝑚+𝑠) = ℎ, Bob conclui que está tudo certo.

```
h = rsa.compute_hash(s + message, 'SHA-1')
print(h)
if (h == hash):
  print(True)
else: 
  print(False)
```
