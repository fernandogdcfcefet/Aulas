Bob e Alice possuem uma chave pública e uma privada cada um, que são únicas

```
import rsa
```

```
(bob_pub, bob_priv) = rsa.newkeys(512)
(alice_pub, alice_priv) = rsa.newkeys(512)
```

Para assinar uma mensagem 𝑚, Bob usa sua chave privada

```
message = 'Querida Alice. Como eu sinto sua falta'.encode()
signature = rsa.sign(message, bob_priv, 'SHA-1')
```

Alice pega a chave pública de Bob e aplica a assinatura digital associada ao documento 𝑚

```
rsa.verify(message, signature, bob_pub)
```
