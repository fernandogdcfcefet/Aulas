```python
cont = 0
recall = 0
acuracia = 0
prev = 1
fpositive = 0
fnegative = 0
tpositive = 0
tnegative = 0

if(bd1[bd1['is_attributed']==1].shape[0]*1.0/bd1.shape[0] < 0.5):
    prev = 0

for i in range(bd1.shape[0]):
    if(bd1.is_attributed[i] == prev):
        cont = cont + 1

acuracia = cont*1.0/bd1.shape[0]
precision = acuracia
recall = 1

fpositive = bd1.shape[0] - cont
tpositive = cont
tnegative = 0
fnegative = 0

if(prev == 0):
    fpositive = 0
    tpositive = 0
    tnegative = cont
    fnegative = bd1.shape[0] - cont
    precision = 0
    recall = 0

print 'acurácia:', acuracia
print 'precisão:', precision
print 'revocação:', recall
print 'true positive:', tpositive
print 'false positive:', fpositive
print 'true negative:', tnegative
print 'false negative:', fnegative
```

# 1-hot encoding 

```python
bd1.head()
```

```python
ip = 1
app = 1
device = 1
os = 1
channel = 1

def ips(e_ip):
    if e_ip == ip:
        return 1
    return 0

def apps(e_app):
    if e_app == app:
        return 1
    return 0

def devices(e_device):
    if e_device == device:
        return 1
    return 0

def oss(e_os):
    if e_os == os:
        return 1
    return 0

def channels(e_channel):
    if e_channel == channel:
        return 1
    return 0

```

```python

cont = 0
for i in bd1.channel.unique():
    channel = i
    bd1['channel'+str(cont)] = bd1.channel.apply(channels)
    cont = cont + 1
```

```python

    
cont = 0
for i in bd1.os.unique():
    os = i
    bd1['os'+str(cont)] = bd1.os.apply(oss)
    cont = cont + 1
```

```python

    
cont = 0
for i in bd1.device.unique():
    device = i
    bd1['device'+str(cont)] = bd1.device.apply(devices)
    cont = cont + 1
```

```python

cont = 0
for i in bd1.app.unique():
    app = i
    bd1['app'+str(cont)] = bd1.app.apply(apps)
    cont = cont + 1
```

```python
cont = 0
for i in bd1.ip.unique():
    ip = i
    if(bd1[bd1['ip']==ip].shape[0]>1):
        bd1['ip'+str(cont)] = bd1.ip.apply(ips)
        cont = cont + 1
    
```

```python
dia = 6

def days(e_dia):
    if int(e_dia.split("-")[2].split(" ")[0]) == dia:
        return 1
    return 0

cont = 0
for i in range(6,10):
    dia = i
    bd1['dia'+str(cont)] = bd1.click_time.apply(days)
    cont = cont + 1
    
```
